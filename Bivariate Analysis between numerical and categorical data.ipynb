```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x = 'class', y = 'age', data = df)
```

```python
sns.barplot(x = 'class', y = 'fare', data = df, estimator = 'sum')
```

```python
sns.barplot(x = 'class', y = 'fare', data = df, estimator = 'mean', hue = 'sex')
```

# Boxplot

```python
sns.boxplot(x = 'sex', y = 'age', data = df)
```

```python
sns.boxplot(x = 'sex', y = 'age', data = df, hue = 'class')
```

```python
sns.violinplot(x = 'sex', y = 'age', data = df, hue = 'class')
```

```python
sns.displot(df[df['class'] == 'Second']['age'])
```
