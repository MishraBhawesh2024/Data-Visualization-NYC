```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('AB_NYC_2019.csv')
```

```python
# A histogram across a categorical variable, as opposed to a quantitative variable, is countplot
sns.countplot(x = data['room_type'])
```

```python
sns.countplot(x = data['neighbourhood_group'])
```

```python
# now same plot above can be further refined to show different kind of room_type in each neighborhood

sns.countplot(data = data, x = 'neighbourhood_group', hue = 'room_type')
```

```python
# Now same thing on pie-charts
```

```python
room_type_counts = data['room_type'].value_counts()
```

```python
room_type_counts.plot(kind = 'pie', autopct = '%.2f')
```

```python
data['availability_365'].value_counts().plot(kind = 'hist', bins = 20)
```

# Now Visualizations of NUMERICAL DATA
